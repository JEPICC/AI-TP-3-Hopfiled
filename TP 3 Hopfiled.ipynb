<h3>Funciones Aprendizaje</h3>

```python
def discretizar(n):
    if n < 0.0:
        return -1.0
    else:
        return 1.0


def multiplicacion_vectores(patrones_ent):
    respuesta = np.zeros((100,100))

    for patron in patrones_ent:
        auxiliar = np.zeros((100,100))
        for i in range(100):
            for j in range(100):
                auxiliar[i][j] = patron[i]* patron[j]
        
        respuesta = respuesta + (auxiliar - np.identity(100))
    return respuesta


pesos = multiplicacion_vectores(patrones)
```

<h3>Patrón de Entrada</h3>

```python
foto1 = np.loadtxt('Foto2.txt')
foto_patron = np.reshape(foto1, 100)
plt.figure(figsize=(10,10))
plt.imshow(np.split(foto_patron, 10))
plt.show
```

<h3>Recuperación</h3>

```python
estimacion = np.dot(foto_patron, pesos)

for i in range(5):
    auxi = discretizar(estimacion.all())
    estimacion = np.dot(auxi, pesos)


estimacion = np.dot(foto_patron, pesos)
grafico_array = []
for item in estimacion:
    grafico_array.append(discretizar(item))

grafico_multi_array = np.array(grafico_array)
np.reshape(grafico_multi_array, (10,10))
plt.figure(figsize=(10,10))
plt.imshow(np.split(grafico_multi_array, 10))
plt.show
```
